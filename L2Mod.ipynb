```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_segmented = pd.read_csv('/Users/renaaaa/Downloads/Segmented_Dialogue.csv' )
data_segmented.head()
```

```python
# Calculate basic statistics

# 1. Number of unique turns, segments, speakers, and roles
num_turns = data_segmented['turn_index'].nunique()
num_segments = data_segmented['segment'].nunique()
num_speakers = data_segmented['speaker'].nunique()
num_roles = data_segmented['role'].nunique()

# 2. Text length statistics
data_segmented['text_length'] = data_segmented['text'].apply(len)  # Length of each text in characters
text_length_stats = data_segmented['text_length'].describe()  # Summary stats for text length

# 3. Frequency distribution of speakers and roles
speaker_counts = data_segmented['speaker'].value_counts()
role_counts = data_segmented['role'].value_counts()

# Plotting

# Plot 1: Histogram of text lengths
plt.figure(figsize=(10, 6))
plt.hist(data_segmented['text_length'], bins=20, edgecolor='black')
plt.title("Distribution of Text Lengths (in Characters)")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.show()

# Plot 2: Bar plot of speaker frequencies
plt.figure(figsize=(10, 6))
speaker_counts.plot(kind='bar')
plt.title("Frequency of Each Speaker")
plt.xlabel("Speaker")
plt.ylabel("Count")
plt.show()

# Plot 3: Bar plot of role frequencies
plt.figure(figsize=(8, 6))
role_counts.plot(kind='bar', color='skyblue')
plt.title("Frequency of Each Role")
plt.xlabel("Role")
plt.ylabel("Count")
plt.show()

# Display calculated statistics for further analysis
text_length_stats, num_turns, num_segments, num_speakers, num_roles
```

```python
# For each segmenation

data_with = pd.read_csv('/Users/renaaaa/Downloads/First_Dialogue_Segmented_Scores.csv')
data_without = pd.read_csv('/Users/renaaaa/Downloads/Second_Dialogue_Segmented_Scores.csv')
```

```python
# Plot function to visualize interactivity aspect scores for segmented dialogues
def plot_segmented_scores(segmented_scores_df, title):
    # Plot each interactivity aspect
    plt.figure(figsize=(12, 8))
    
    # Plot Topic Management scores
    plt.subplot(2, 2, 1)
    plt.plot(segmented_scores_df['Segment ID'], segmented_scores_df['Topic Management'], marker='o', linestyle='-', color='b')
    plt.title('Topic Management')
    plt.xlabel('Segment ID')
    plt.ylabel('Score')
    plt.ylim(0, 5)
    
    # Plot Tone Appropriateness scores
    plt.subplot(2, 2, 2)
    plt.plot(segmented_scores_df['Segment ID'], segmented_scores_df['Tone Appropriateness'], marker='o', linestyle='-', color='g')
    plt.title('Tone Appropriateness')
    plt.xlabel('Segment ID')
    plt.ylabel('Score')
    plt.ylim(0, 5)
    
    # Plot Conversation Opening scores
    plt.subplot(2, 2, 3)
    plt.plot(segmented_scores_df['Segment ID'], segmented_scores_df['Conversation Opening'], marker='o', linestyle='-', color='r')
    plt.title('Conversation Opening')
    plt.xlabel('Segment ID')
    plt.ylabel('Score')
    plt.ylim(0, 5)
    
    # Plot Conversation Closing scores
    plt.subplot(2, 2, 4)
    plt.plot(segmented_scores_df['Segment ID'], segmented_scores_df['Conversation Closing'], marker='o', linestyle='-', color='purple')
    plt.title('Conversation Closing')
    plt.xlabel('Segment ID')
    plt.ylabel('Score')
    plt.ylim(0, 5)
    
    # Overall title for the plot
    plt.suptitle(title, fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Plot results for the first dialogue
plot_segmented_scores(data_with, "Interactivity Aspect Scores - With Moderator")

# Plot results for the second dialogue
plot_segmented_scores(data_without, "Interactivity Aspect Scores - Without Moderator")
```
